__Задание 2 (1 балл):__ Опишите класс комплексных чисел. У пользователя должна быть возможность создать его объект на основе числа и в алгебраической форме, и в полярной. Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
import math
class ComplexNumber:
    #number_type - форма числа - полярная или алгебраическая
    #first, second - два числа, которые будут задавать комплексное в каждом из возможных представлений числа.
    def __init__(self, number_type, first, second):
        if number_type == 'algebraic':
            if not (isinstance(first, (int, float)) and isinstance(second, (int, float))):
                raise ValueError('First and second part of number should be numbers')
            self._real_part = first
            self._complex_part = second
            self._module = (first ** 2 + second ** 2) ** 0.5
            if first == 0:
                if second > 0:
                    self._angle = math.pi / 2
                elif second < 0:
                    self._angle = -math.pi / 2
                else: 
                    self._angle = 0
            else:
                self._angle = math.atan2(second, first)        
        elif number_type == 'polar':
            if not (isinstance(first, (int, float)) and isinstance(second, (int, float))):
                raise ValueError('First and second part of number should be numbers')
            if first < 0:
                raise ValueError('It should be positive')
            self._real_part = first*math.cos(second)
            self._complex_part = first*math.sin(second)
            self._module = first
            self._angle = second
        else:
            raise ValueError('There is no such type of complex number representation')
    #реализуем сложение двух чисел    
    def __add__(number1, number2):
        add_real_part = number1._real_part + number2._real_part
        add_complex_part = number1._complex_part + number2._complex_part
        add_res = ComplexNumber('algebraic', add_real_part, add_complex_part)
        return add_res
    #реализуем вычитание двух чисел
    def __sub__(number1, number2):
        sub_real_part = number1._real_part - number2._real_part
        sub_complex_part = number1._complex_part - number2._complex_part
        sub_res = ComplexNumber('algebraic', sub_real_part, sub_complex_part)
        return sub_res
    #реализуем умножение двух чисел
    def __mul__(number1, number2):
        mul_real_part = number1._real_part * number2._real_part - number1._complex_part * number2._complex_part
        mul_complex_part = number1._real_part * number2._complex_part + number1._complex_part * number2._real_part
        mul_res = ComplexNumber('algebraic', mul_real_part, mul_complex_part)
        return mul_res
    #реализуем деление двух чисел
    def __truediv__(number1, number2):
        if number2._real_part == 0 and number2._complex_part == 0:
            raise ZeroDivisionError('Division by zero')
        div_real_part = (number1._real_part * number2._real_part + number1._complex_part * number2._complex_part) / (number2._real_part ** 2 + number2._complex_part ** 2)
        div_complex_part = (number1._complex_part * number2._real_part - number1._real_part * number2._complex_part) / (number2._real_part ** 2 + number2._complex_part ** 2)
        div_res = ComplexNumber('algebraic', div_real_part, div_complex_part)
        return div_res
    #реализуем получение числа в алг.форме
    def receive_alg(self):
        return [self._real_part, self._complex_part]
    #реализуем получение числа в полярной форме
    def receive_polar(self):
        return [self._module, self._angle]
    
````

__Задание 3 (2 балла):__ Опишите класс для векторов в N-мерном пространстве. В качестве основы  используйте список значений координат вектора, задаваемый `list`. Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
class Vector:
    def __init__(self, vector_values_list):
        if not isinstance(vector_values_list, list):
            raise ValueError('Invalid Input: type is not list. Please enter again.')
        for i in vector_values_list:
            if not (isinstance(i, (int, float))):
                raise ValueError('Invalid Input: vector coordinates must be numbers.Please enter again.')
        if len(vector_values_list) == 0:
            raise ValueError('Invalid Input: vector dimension must be a natural number.Please enter again')
        self._components = vector_values_list
        self._dim = len(vector_values_list)
    #сложение
    def __add__(self, vector2):
        if self._dim != vector2._dim:
            raise ValueError('Dimensions of vectors are not equal')
        else:
            add_result = Vector(list(map(lambda x: x[0] + x[1], zip(self._components, vector2._components))))
            return add_result
    #вычитание
    def __sub__(self, vector2):
        if self._dim != vector2._dim:
            raise ValueError('Dimensions of vectors are not equal')
        else:
            sub_result = Vector(list(map(lambda x: x[0] - x[1], zip(self._components, vector2._components))))
            return sub_result
    #скалярное произведение    
    def scalar_mul(self, vector2):
        if self._dim != vector2._dim:
            raise ValueError('Dimensions of vectors are not equal')
        else:
            return sum(list(map(lambda x: x[0] * x[1], zip(self._components, vector2._components))))
    #евклидова норма
    def norm(self):
        return sum([i ** 2 for i in self._components]) ** 0.5
    #косинус угла
    def cos_angle(self, vector2):
        if self._dim != vector2._dim:
            raise ValueError('Dimensions of vectors are not equal')
        if self.norm() == 0 or vector2.norm() == 0:
            raise ValueError("Length of vector is 0. Angle is not defined")
        else:
            return self.scalar_mul(vector2) / (self.norm() * vector2.norm())     
        
    def __repr__(self):
        return '{}'.format(self._components)
````

````python
#пользовался материалами с pythonworld для созда
def calls_counter(func):
    def func_counter(*args, **kw):
        func_counter.count += 1
        result = func(*args, **kw)
        print('This function was called {} time'.format(func_counter.count))
        return result
    func_counter.count = 0
    return func_counter
````

````python
#ТЕСТЫ
@calls_counter
def func1():
    for i in range(100):
        pass
@calls_counter
def func2():
    for i in range(100):
        pass
@calls_counter
def func3():
    for i in range(100):
        pass
func1()
func2()
func3()
func3()
func2()
func1()
func2()
func2()
func2()
func2()
func2()
func2()
func2()
````

````python
import types
import functools
import copy
````

````python
def copy_func(f):
    g = types.FunctionType(f.__code__, f.__globals__, name=f.__name__, argdefs=f.__defaults__, closure=f.__closure__)
    g.__kwdefaults__ = f.__kwdefaults__
    return g
````

````python
def copy_class_attrs(cls):
    def __copy_class_attrs(main_cls):
        for e in list(cls.__dict__):
            if e[0:2] != '__' != '__': #смотрим, начинается ли с __ 
                if callable(getattr(cls, e)):
                    if e[0 : len(cls.__name__) + 3] == '_' + cls.__name__ + '__':
                        if hasattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):]):
                            setattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):] + '_new', copy_func(getattr(cls, e)))
                        else:
                            setattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):], copy_func(getattr(cls, e)))
                    else:
                        if hasattr(main_cls, e):
                            setattr(main_cls, e + '_new', copy_func(getattr(cls, e)))
                        else:
                            setattr(main_cls, e, copy_func(getattr(cls, e)))
                else:
                    if e[0 : len(cls.__name__) + 3] == '_' + cls.__name__ + '__':
                        if hasattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):]):
                            setattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):] + '_new', copy.deepcopy(getattr(cls, e)))
                        else:
                            setattr(main_cls, '_' + main_cls.__name__ + e[e.find(cls.__name__) + len(cls.__name__):], copy.deepcopy(getattr(cls, e)))
                    else:
                        if hasattr(main_cls, e):
                            setattr(main_cls, e + '_new', copy.deepcopy(getattr(cls, e)))
                        else:
                            setattr(main_cls, e, copy.deepcopy(getattr(cls, e)))
        return main_cls
    return __copy_class_attrs
                
````

Обе процедуры должны быть реализованы с помощью шаблона "примесь" (Mixin), т.е. указанные функциональности должны быть описаны в специализированных классах.

В реализации математических операций запрещается пользоваться любыми функциями, требующими использования оператора `import`.

````python
class Matrix:
    def __init__(self, type_of_matrix, arg1, arg2 = None, arg3 = None):
            #сначала первым способом через списки
            if type_of_matrix == 'type 1':
                if not isinstance(arg1, list):
                    raise ValueError('Invalid input: type is not list')
                if not arg1:
                    raise ValueError('Invalid input: list is empty')
                if not isinstance(arg1[0], list):
                    arg = arg1
                    arg1 = []
                    arg1.append(arg)
                len_line = len(arg1[0])
                for i in arg1:
                    if not isinstance(i, list):
                        raise ValueError('Invalid input: elements of list are not lists')
                    for k in i:
                        if not (isinstance(k, (int, float))):
                            raise ValueError('Invalid Input: elements of matrix must be numbers')
                if not (arg2 is None and arg3 is None):
                    raise ValueError('Invalid input: extra data')
                self._dim1 = len(arg1)
                self._dim2 = len_line
                self._matrix = arg1
            #теперь явно через 3 числа
            if type_of_matrix == 'type 2':
                if arg1 is None or arg2 is None:
                    raise ValueError('Invalid input: not enough data')
                if not (isinstance(arg1, int) and isinstance(arg2, int)):
                    raise ValueError('Invalid input: matrix sizes should be natural numbers')
                if not (arg1 > 0 and arg2 > 0):
                    raise ValueError('Invalid input: matrix sizes should be natural numbers')
                self._dim1 = arg1
                self._dim2 = arg2
                if arg3 is None:
                    arg3 = 0
                else:
                    if not (isinstance(arg3, (int, float))):
                        raise ValueError('Invalid input: elements of matrix should be numbers')
                self._matrix = [[arg3 for i in range(self._dim2)] for j in range(self._dim1)]
            self._slice = 0
            self._number = 0
    def __repr__(self):
                return '{}'.format(self._matrix)
            
    def __setitem__(self, index, value):
        if not isinstance(index, int):
            raise ValueError('Invalid Input: index must be integer number')
        if not (isinstance(value, int) or isinstance(value, float)):
            raise ValueError('Invalid Input: value must be number')
        self.old_matrix._matrix[self.index][index] = value

    def __getitem__(self, index):
        if not (isinstance(index, int)  or index == slice(None, None, None)):
            raise ValueError('Invalid index')
        if index != slice(None, None, None):
            if self._number == 0 and self._slice == 0:
                result =  Matrix('type 1', self._matrix[index])
                setattr(result, 'old_matrix', self)
                setattr(result, 'index', index)
                result._number = 1
                return result
            elif self._number == 1:
                self._number = 0
                return self._matrix[0][index]
            elif self._slice == 1:
                self._slice = 0
                lst = []
                for i in self._matrix:
                    lst.append([i[index]])
                return Matrix('type 1', lst)   
        else:
            if self._slice == 1:
                raise ValueError('Invalid index')
            if self._number == 0:
                self._slice = 1
            else:
                self._number = 0
        return self

    #реализуем сложение двух матриц
    def __add__(self, matrix):
        if self._dim1 != matrix._dim1 or self._dim2 != matrix._dim2:
            raise ValueError('Different dimensions of matrixes')
        result = []
        for i in range(self._dim1):
            result.append(list(map(lambda x: x[0] + x[1], zip(self._matrix[i], matrix._matrix[i]))))
        return Matrix('type 1', result)
    #реализуем разность двух матриц
    def __sub__(self, matrix):
        if self._dim1 != matrix._dim1 or self._dim2 != matrix._dim2:
            raise ValueError('Different dimensions of matrixes')
        result = []
        for i in range(self._dim1):
            result.append(list(map(lambda x: x[0] - x[1], zip(self._matrix[i], matrix._matrix[i]))))
        return Matrix('type 2', result)
    #реализуем умножение матрицы на число
    def __mul__(self, number):
        result = []
        for i in range(self._dim1):
            result.append(list(map(lambda x: x * number, self._matrix[i])))
        return Matrix('type 1', result)
    #реализуем деление матрицы на число
    def __truediv__(self, number):
        if number == 0:
            raise ZeroDivisionError('Division by zero')
        matrix = []
        for line in self._matrix:
            lst = []
            for i in line:
                lst.append(i / number)
            matrix.append(lst)
        return Matrix('type 1', matrix)

    def dot(self, matrix):
        if self._dim2 != matrix._dim1:
            raise ValueError('Invalid matrix dimensions')
        result = []
        for i in range(self._dim1):
            lst = []
            for j in range(matrix._dim2):
                el = 0
                for k in range(self._dim2):
                    el += (self[i][:])[0][k] * (matrix[:][j])[k][0]
                lst.append(el)
            result.append(lst)
        return Matrix('type 1', result)
    #реализуем транспонирование
    def transpose(self):
        res = Matrix('type 2', self._dim2, self._dim1)
        for i in range(self._dim1):
            for j in range(self._dim2):
                res[j][i] = self[i][j]
        return res  
    #реализуем поиск следа матрицы    
    def trace(self):
        if self._dim1 != self._dim2:
            raise ValueError('Invalid input: matrix is not squared')
        summa = 0
        for i in range(self._dim1):
            summa += self[i][i]
        return summa
    #реализуем поиск детерминанта матрицы
    def determinant(self):
        if self._dim1 != self._dim2:
            print('Matrix must be square')
            return None
        if self._dim1 != 1:
            summa = 0
            for i in range(self._dim1):
                res = [] 
                for l in range(1, self._dim1):
                    lst = []
                    j = 0
                    for j in range(self._dim1):
                        if j != i:
                            lst.append(self[l][:][0][j])
                    res.append(lst)
                Det = Matrix('type 1', res)
                summa += ((-1) ** i) * self[0][i] * Det.determinant()
            return summa
        else:
            return self[0][0]

    def copy(self):
        import copy
        lst = copy.deepcopy(self._matrix)
        return Matrix('type 1', lst)

    def __next__(self):
        index = self._index
        if self._dim1 == 1:
            self._index = index + 1
            if self._index >= self._dim2:
                raise StopIteration
            return self[0][self._index]
        elif self._dim2 == 1:
            self._index = index + 1
            if self._index >= self._dim1:
                raise StopIteration
            return self[self._index][0]

    def __iter__(self):
        self._index = -1
        if self._dim1 != 1 and self._dim2 != 1:
            raise TypeError('Object is not iterable')
        return self

class Mixin:
    def save_bin_file(self, file):
        import pickle
        with open(file, 'wb') as fout:
            pickle.dump(self, fout)
            
    def read_bin_file(file):
        import pickle
        with open(file, 'rb') as fin:
            a = pickle.load(fin)
            if a.__class__ != Matrix:
                raise ValueError('Object in file does not belong to class Matrix')
            return a
    
    def save_text_file(self, file):
        with open(file, 'w')as fout:
            for j in range(self._dim1):
                fout.write(str(self[j][:]) + '\n')
    
    def read_text_file(file):
        with open(file, 'r') as fin:
            matrix = []
            for line in fin:
                lst = []
                new_line = line[2:len(line)-3].split(',')
                for j in range(len(new_line)):
                    lst.append(float(new_line[j]))
                matrix.append(lst)
            return Matrix('type 1', matrix)

class Matrix_Mixin(Matrix, Mixin):
    def __init__(self, value):
        super().__init__(value)
````
